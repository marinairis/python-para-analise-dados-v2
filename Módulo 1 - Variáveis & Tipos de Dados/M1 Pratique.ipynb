4º Exercício: Booleanos

```python
# PARTE 1

a = False
b = True
x = not a & b

print(x)

# PARTE 2
news = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

start_selic = news.find('a ') + 2
end_selic = news.find('%')
value_selic = news[start_selic:end_selic]

start_year = news.find('em ') + 3
end_year = news.find(' anos')
year = news[start_year:end_year]

print("Valor da Selic:", value_selic)
print("Ano:", year)
```
